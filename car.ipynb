```python
# 완벽한 코드

# 시간은 나중에 조정

if btn.clicked:
    # straight
    if turn == 0 and turn ==100:

        motor0 = bundle.motors[0]
        motor0.speed = 30
        time.sleep(1)

        motor1 = bundle.motors[1]
        motor1.speed = 30
        time.sleep(1)

        motor2 = bundle.motors[2]
        motor2.speed = 30
        time.sleep(1)

        motor3 = bundle.motors[3]
        motor3.speed = 30
        time.sleep(1)

    # right
    if 0 < turn and turn < 50:

        motor0 = bundle.motors[0]
        motor0.speed = 50
        time.sleep(1)

        motor1 = bundle.motors[1]
        motor1.speed = 50
        time.sleep(1)

        motor2 = bundle.motors[2]
        motor2.speed = 30
        time.sleep(1)

        motor3 = bundle.motors[3]
        motor3.speed = 30
        time.sleep(1)

    # left

    if 50 < turn and turn < 100:

        motor0 = bundle.motors[0]
        motor0.speed = 30
        time.sleep(1)

        motor1 = bundle.motors[1]
        motor1.speed = 30
        time.sleep(1)

        motor2 = bundle.motors[2]
        motor2.speed = 50
        time.sleep(1)

        motor3 = bundle.motors[3]
        motor3.speed = 50
        time.sleep(1)

        # back

    if turn == 50:

        motor0 = bundle.motors[0]
        motor0.speed = -30
        time.sleep(1)

        motor1 = bundle.motors[1]
        motor1.speed = -30
        time.sleep(1)

        motor2 = bundle.motors[2]
        motor2.speed = -30
        time.sleep(1)

        motor3 = bundle.motors[3]
        motor3.speed = -30 
        time.sleep(1)


if btn.double_clicked:
    print("Stop!!")
    motor0.speed = 0
    motor1.speed = 0
    motor2.speed = 0
    motor3.speed = 0
```
